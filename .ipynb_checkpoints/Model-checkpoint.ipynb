```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
def load_image_pairs(noisy_dir, clean_dir, img_size=(64, 64)):
    image_names = sorted(os.listdir(noisy_dir))
    noisy_images = []
    clean_images = []

    for name in image_names:
        noisy_path = os.path.join(noisy_dir, name)
        clean_path = os.path.join(clean_dir, name)

        noisy_img = load_img(noisy_path, target_size=img_size)
        clean_img = load_img(clean_path, target_size=img_size)

        noisy_img = img_to_array(noisy_img) / 255.0
        clean_img = img_to_array(clean_img) / 255.0

        noisy_images.append(noisy_img)
        clean_images.append(clean_img)

    return np.array(noisy_images), np.array(clean_images)

# Load data
x_train, y_train = load_image_pairs("E:/Code/project/Chana/Code/Image Denoising via Outlier Detection using CNN-U-Net Hybrid Model/images/noisy/train", "E:/Code/project/Chana/Code/Image Denoising via Outlier Detection using CNN-U-Net Hybrid Model/images/clean/train")
x_val, y_val = load_image_pairs("E:/Code/project/Chana/Code/Image Denoising via Outlier Detection using CNN-U-Net Hybrid Model/images/noisy/val", "E:/Code/project/Chana/Code/Image Denoising via Outlier Detection using CNN-U-Net Hybrid Model/images/clean/val")
x_test, y_test = load_image_pairs("E:/Code/project/Chana/Code/Image Denoising via Outlier Detection using CNN-U-Net Hybrid Model/images/noisy/test", "E:/Code/project/Chana/Code/Image Denoising via Outlier Detection using CNN-U-Net Hybrid Model/images/clean/test")
```

```python
def conv_block(x, filters):
    x = layers.Conv2D(filters, 3, padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Conv2D(filters, 3, padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    return x

def encoder_block(x, filters):
    c = conv_block(x, filters)
    p = layers.MaxPooling2D((2, 2))(c)
    return c, p

def decoder_block(x, skip, filters):
    x = layers.Conv2DTranspose(filters, (2, 2), strides=2, padding='same')(x)
    x = layers.concatenate([x, skip])
    x = conv_block(x, filters)
    return x

def build_unet(input_shape=(64, 64, 3)):
    inputs = layers.Input(shape=input_shape)

    c1, p1 = encoder_block(inputs, 64)
    c2, p2 = encoder_block(p1, 128)
    c3, p3 = encoder_block(p2, 256)

    bn = conv_block(p3, 512)

    d3 = decoder_block(bn, c3, 256)
    d2 = decoder_block(d3, c2, 128)
    d1 = decoder_block(d2, c1, 64)

    outputs = layers.Conv2D(3, (1, 1), activation='sigmoid')(d1)

    return models.Model(inputs, outputs)

model = build_unet()
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.summary()
```

```python
history = model.fit(
    x_train, y_train,
    validation_data=(x_val, y_val),
    epochs=10,
    batch_size=8
)
```

```python
# Test one batch
preds = model.predict(x_test[:5])

for i in range(5):
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 3, 1)
    plt.imshow(x_test[i])
    plt.title("Noisy")

    plt.subplot(1, 3, 2)
    plt.imshow(preds[i])
    plt.title("Denoised")

    plt.subplot(1, 3, 3)
    plt.imshow(y_test[i])
    plt.title("Ground Truth")

    plt.tight_layout()
    plt.show()
```
